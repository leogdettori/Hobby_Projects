```python
#This script downloads a single pokemon data entry from the Serebii website using the function wget


import wget

#Output Folder
output = 'C:/Users/leogd/swsh_pokedex_Serebii/'

#Serebii URL
url = "https://www.serebii.net/pokedex-swsh/"

#Curent pokemon (ready to set up a loop with a list)
current_pokemon = "ursaring"

#Downloads the data for the current pokemon using specified URL and ouput location
wget.download(url+current_pokemon, out=output+current_pokemon+".txt")
```

```python
#This script opens up the txt file with all pokemon names acquired from a serebii pokemon data entry
#and organizes all pokemon names and numbers in a python list (number_name)

dex_file = 'C:/Users/leogd/swsh_pokedex.txt'

## ============================================================================================== ##

# Reads the data
f = open(dex_file)

#creates list
dex_list = []

#Starts the loop through the file. Line by line.
for line in f:
    #Splits line and replaces space with underscore to extract pokemon number_name    
    this_line = line.split('>')[1].split('<')[0].replace(' ','_')
    
    #Dealing with the nidoran problem
    if 'Nidoran' in this_line:
        #print(this_line)        
        #For female nidoran
        if '29' in this_line:
            this_line = '029_Nidoranf'
        #For male nidoran
        if '32' in this_line:
            this_line = '032_Nidoranm'
    
    #Dealing with the Mr. Mime and Mime Jr. problem
    if '122' in this_line:
        this_line = '122_Mr.Mime'
        #print(this_line)
    if '439' in this_line:
        this_line = '439_MimeJr.'
        #print(this_line)       
        
    #Dealing with the Type: Null problem
    if '772' in this_line:
        this_line = '772_Type:Null'
        #print(this_line)
    ##Since : cannot be used in a file's name, Type:Null had to be handled separately
        
    #Dealing with the Tapu problems
    if '785' in this_line:
        this_line = '785_TapuKoko'
        #print(this_line)
    if '786' in this_line:
        this_line = '786_TapuLele'
        #print(this_line)   
    if '787' in this_line:
        this_line = '787_TapuBulu'
        #print(this_line)
    if '788' in this_line:
        this_line = '788_TapuFini'
        #print(this_line)
        
    #Dealing with the Mr. Rime problem
    if '866' in this_line:
        this_line = '866_Mr.Rime'
        #print(this_line)                
        
    
    #Adds info to dex_list
    dex_list.append(this_line)    
    #print(this_line)

f.close()

print(dex_list)
```

```python
#This script downloads all pokemon data entries from the Serebii website 
#using the function wget and a provided poke_list (number_name)

#Poke_list (number_name)
poke_list = dex_list

#Output Folder
output = 'C:/Users/leogd/swsh_pokedex_Serebii/'

#Serebii URL
url = "https://www.serebii.net/pokedex-swsh/"

#dex generation name:
dex_gen = "swsh"


## ============================================================================================== ##

import wget


#Starts the loop trhough the poke_list
for pokemon in poke_list:
    
    #Extracts pokemon name and number from current pokemon entry in poke_list
    #sword and shield dex works with name
    current_name = pokemon.split('_')[1].lower()  #converts name to all lowercase
    #images and all the other dexes work with numbers
    current_number = pokemon.split('_')[0]
    
    #Taking care of Type:Null separately since : cannot be used in a file's name
    if int(current_number) == 772:
        #Downloads the data for the current pokemon using specified URL and ouput location
        wget.download(url+current_name, out=output+"772_TypeNull"+".pokedex"+dex_gen)
    
    #For everyone else
    else:
        #Downloads the data for the current pokemon using specified URL and ouput location
        wget.download(url+current_name, out=output+pokemon+".pokedex"+dex_gen)
```
